## Preparación de datos para región Libertador Bernardo O'Higgins

```python

terrenos_df = gpd.read_file('predios_agricola_ohiggins.geojson')

terr_jsons = []

terrenos_df
for _, row in terrenos_df.iterrows():
    obj = {"cod_comuna": int(row.comudere), 
    "comuna": row.desccomu,
    "provincia": str(row.provdere), 
    "area_ha": row.area_hz,
    "primera_especie": "algo"}
    terr_jsons.append(obj)

with open("../inputs/terrenos_oh.json", "w") as f:
    import json
    json.dump(terr_jsons, f, indent=4)
    
```

```python
dias_to_int = {
    "Todos los días": [0,1,2,3,4,5,6],
    "Lunes": 0,
    "Martes": 1,
    "Miércoles": 2,
    "Jueves": 3,
    "Viernes": 4,
    "Sábado": 5,
    "Domingo": 6,
}

def get_dias(dias) -> list[int]: 
    vals = dias.split(', ')
    days = []
    for val in vals:
        if val == "Todos los días":
            return [0,1,2,3,4,5,6]
        day = dias_to_int.get(val, None)
        if day is None:
            val = val.split(' y ')
            for el in val:
                days.append(dias_to_int[el])
            continue
        days.append(day)

    print(days)
    return days
```

```python
ferias_df = pd.read_excel("Ferias_Pais_PowerBI.xlsx").query("REGIÓN == 6")
ferias_df = ferias_df.rename(columns={"N°": "nums"})
ferias_df["Dias_de_Postura"].unique()
ferias_filtered = ferias_df.query("Dias_de_Postura in ('Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo',"
       "'Jueves y Domingo', 'Todos los días', 'Martes, Viernes y Sábado',"
       "'Viernes, Sábado y Domingo', " 
       "'Martes y Viernes', 'Miércoles y Domingo', 'Miércoles y Viernes',"
       "'Lunes, Martes, Jueves y Viernes', 'Miércoles y Sábado') and nums != 1")

ferias_filtered["dias"] = ferias_filtered["Dias_de_Postura"].apply(get_dias)
ferias_filtered
```

```python
objs = []
total_puestos = 0
for _, row in ferias_filtered.iterrows():
    
    obj = {
        "cod_comuna": str(row.ID),
        "comuna": row.COMUNA,
        "dias": row.dias,
        "cantidad_puestos": row.nums
    }
    total_puestos += row.nums
    objs.append(obj)

```

```python

with open("../inputs/ferias_oh.json", "w", encoding="utf-8") as f:
    import json
    json.dump(objs, f, indent=4)
    
```

```python
spi_df = pd.read_csv('SPI.csv')
spi_df
aux = {}
arr = []
import json
for index, row in spi_df.iterrows():
    if row["SPI_rainfall.fitted"] >= 0:
        arr.append(0)
    elif row["SPI_rainfall.fitted"] >= -1:
        arr.append(1)
    elif row["SPI_rainfall.fitted"] >= -2:
        arr.append(2)
    else:
        arr.append(3)
aux["levels"] = arr

f = open("./spi_levels_oh.json", "w")
f.write(json.dumps(aux, indent=4))
f.close()
```
